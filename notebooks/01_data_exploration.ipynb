```python
import pickle
import numpy as np

# Load the dataset
with open('../data/RML2016.10a_dict.pkl', 'rb') as f:
    data = pickle.load(f, encoding='latin1')

# See what's inside
print(type(data))
print(data.keys())
```

```python
# Look at one sample
sample = data[('QPSK', 2)]
print("Shape:", sample.shape)
```

```python
sample = data[('QPSK', 2)]

print("Full shape:", sample.shape)
print("One signal (2, 128):\n", sample[0])
print("\nJust I channel (128 values):", sample[0][0])
print("Just Q channel (128 values):", sample[0][1])
```

```python
import matplotlib.pyplot as plt

sample = data[('QPSK', 2)]

plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(sample[0][0], label='I channel')
plt.plot(sample[0][1], label='Q channel')
plt.title('QPSK at SNR=2dB (time domain)')
plt.legend()

plt.subplot(1, 2, 2)
plt.scatter(sample[:, 0, :], sample[:, 1, :], alpha=0.1, s=1)
plt.title('QPSK Constellation Diagram')
plt.xlabel('I')
plt.ylabel('Q')

plt.tight_layout()
plt.show()
```
